```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def create_array(n):
    l = [int(x) for x in np.zeros(n)]
    l[n//2] = 1
    return l


def add_array(arr):
    i = 0
    new_array= []
    for i in range(len(arr)):
        if i == len(arr) - 1: 
            new_array.append(0)
            break

        #000, 101, 110, 111 -> 0
        elif any([
            all([arr[i-1] == 0, arr[i] == 0, arr[i+1] == 0])
            , all([arr[i-1] == 1, arr[i] == 0, arr[i+1] == 1])
            , all([arr[i-1] == 1, arr[i] == 1, arr[i+1] == 0])
            , all([arr[i-1] == 1, arr[i] == 1, arr[i+1] == 1]) 
         ]):
            new_array.append(0)
        else:
            new_array.append(1)
            
    return new_array
```

```python
init_arr = create_array(1000)

a = []
x = []
for i in range(1000):
    if i == 0:
        a.append(init_arr)
    elif i == 1:
        x = add_array(init_arr)
        a.append(x)
    elif i > 1:
        x = add_array(x)
        a.append(x)
        
y=np.array([np.array(xi) for xi in a])

plt.figure(figsize = (20,40))

plt.imshow(y, interpolation='nearest')
plt.show()
```
